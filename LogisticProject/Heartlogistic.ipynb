# Train Test Splitting and Scaling

```python
df.head()
```

```python
X = df.drop('target', axis =1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 101)
```

```python
df.shape
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_tarnsform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
log_model = LogisticRegressionCV()
```

```python
log_model.fit(scaled_X_train, y_train)
```

```python
help(LogisticRegressionCV())
```

```python
log_model.Cs_
```

```python
log_model.C_
```

```python
log_model.get_params()
```

```python
log_model.coef_
```

```python
coefs = pd.Series(index = X.columns, data = log_model.coef_[0] )
coefs
```

```python
plt.figure(figsize=(10,6), dpi =150)
coefs = coefs.sort_values()
sns.barplot( x= coefs.index, y = coefs.values)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, plot_confusion_matrix, plot_roc_curve, plot_precision_recall_curve
```

```python
y_pred = log_model.predict(scaled_X_test)
y_pred
```

```python
confusion_matrix(y_pred, y_test)
```

```python
plot_confusion_matrix(log_model, scaled_X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
plot_roc_curve(log_model, scaled_X_test, y_test)
```

```python
plot_precision_recall_curve(log_model, scaled_X_test, y_test)
```
